```python
#importanção do pandas
import pandas as pd
import numpy as np
```

```python
pd.read_csv('saudedosono.csv') # importação do dataframe
df = pd.read_csv('saudedosono.csv') # definição da variavél
df.head(3) # mostrar 3 linhas do shape
```

```python
# renomeando as colunas = realizado criação de uma variavél com um dicionario de mapeamento e incluso para alteração ao
# dataframe df.rename(columns=nova_coluns, inplace=True)

nova_coluns = {
    'ID': 'identificação',
    'Gênero': 'gênero',
    'Idade': 'idade',
    'Ocupação': 'profissão',
    'Duração do sono': 'duração do sono',
    'Qualidade do sono': 'qualidade do sono',
    'Nível de atividade': 'nível de atividade física',
    'Nível de estresse': 'nível de estresse',
    'Categoria BMI': 'categoria imc',
    'Pressão sanguínea': 'pressão sanguínea',
    'Frequência cardíaca': 'frequência cardíaca',
    'Passos diários': 'passos diários',
    'Distúrbio do sono': 'distúrbio do sono'
}

df.rename(columns=nova_coluns, inplace=True)
df.head(3)
```

```python
# Variavél criada para agrupar as categorias profissão e duração do sono e conseguirmos extrair média, mediana e mode
result = df.groupby('profissão')['duração do sono'].agg(['mean', 'median', pd.Series.mode])

print(result)
```

```python
obesos = df.groupby('profissão')['categoria imc'].count()
print(obesos)
```

```python
#Porcentam total das pessoas Engenheiro(a) com sobre peso

engeob = df[(df['profissão'] == 'Engenheiro(a)') & (df['categoria imc'] == 'Sobrepeso')].shape[0] / df[df['profissão'] == 'Engenheiro(a)'].shape[0]
print(engeob)
```
